```python
import numpy as np
from bokeh.plotting import figure
from bokeh.io import push_notebook, show, output_notebook
from bokeh.models.widgets import Div
from bokeh.layouts import row, column
from  ipywidgets import interact
output_notebook()

A = np.array([[1, 0], [0, 1]])
sq = np.array([[0,0],[1,0],[1,1],[0,1]]).T # random square

# plot square before transormations
plt = figure(x_range=[-2, 3], y_range=[-2, 3], 
             plot_width=350, plot_height=350)
r0 = plt.patch(sq[0,:], sq[1,:], alpha=0.15, line_width=1, color="grey", legend="original shape")
r  = plt.patch(sq[0,:], sq[1,:], alpha=0.25, line_width=2, color="blue", legend="transformed shape (A.shape)")

# plot eigen vectors scaled by eigenvalues
eigval, eigvec = np.linalg.eig(A)
r1 = plt.line(np.array([0, eigvec[0,0]])*eigval[0], 
              np.array([0, eigvec[1,0]])*eigval[0], color='red', legend="eigenvectors of A")
r2 = plt.line(np.array([0, eigvec[0,1]])*eigval[1], 
              np.array([0, eigvec[1,1]])*eigval[1], color='red')

div = Div(width=400,
          text=""" """)

def update(a11=1, a12=0, a21=0, a22=1):
    A = np.array([[a11, a12], [a21, a22]])    
    new_sq = A@sq
    eigval, eigvec = np.linalg.eig(A)
    print('eigenvalues: \n {}'.format(eigval))
    print('eigenvectors: \n{}'.format(eigvec))
    
    r.data_source.data['x'] = new_sq[0,:]
    r.data_source.data['y'] = new_sq[1,:]
    
    if np.imag(eigval[0]) == 0:
        r1.data_source.data['x'] = np.array([0, eigvec[0,0]])*eigval[0]
        r1.data_source.data['y'] = np.array([0, eigvec[1,0]])*eigval[0]
    else:
        r1.data_source.data['x'] = np.array([0,0])
        r1.data_source.data['y'] = np.array([0,0])
    if np.imag(eigval[1]) == 0:
        r2.data_source.data['x'] = np.array([0, eigvec[0,1]])*eigval[1]
        r2.data_source.data['y'] = np.array([0, eigvec[1,1]])*eigval[1]
    else:
        r2.data_source.data['x'] = np.array([0,0])
        r2.data_source.data['y'] = np.array([0,0])            
                                        
    push_notebook()

# show plot
show(row(plt, div), notebook_handle=True)
interact(update, a11=(-1.5,1.5,0.1), a12=(-1.5,1.5,0.1), a21=(-1.5,1.5,0.1), a22=(-1.5,1.5,0.1))
    
```
